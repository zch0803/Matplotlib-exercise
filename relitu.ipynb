```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
%matplotlib inline
input = pd.read_csv('GO.csv', sep=',')
data = input.drop('GO result', axis=1)
data2 = pd.DataFrame(data.values, index=input['GO result'].values, columns=data.columns.values)

print(data2)

#x = np.isnan(data)
```

```python
%matplotlib inline
f, ax = plt.subplots(figsize = (4, 10))
sns.heatmap(data2, linewidths = 0.05, ax = ax, cmap=sns.color_palette("Blues"))
ax.text(8,-1,'-log(pvalue)',fontsize=10)
f.savefig('heatmap.jpg', bbox_inches='tight')
```
