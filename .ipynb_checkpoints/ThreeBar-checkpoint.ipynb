```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
```

```python
input = '''393	0	235
1458	1300	959
2398	1899	1472
3084	2158	1676
3411	2003	2101
2791	1481	1311
4274	2794	3569
1426	721	1426'''

celllines = ['MII', 'zygote', 'early', '2cell', '4cell', '8cell', 'ICM', 'mESC']
All = []
Gain = []
Loss = []

input_process = input.split('\n')

for line in input_process:
    data = line.split('\t')
    All.append(int(data[0]))
    Gain.append(int(data[1]))
    Loss.append(int(data[2]))


index = np.arange(8)
#print(index)
bar_width = 0.2
colorBar1 = [[0,0.2,0.8],[0.0,0.4,0.8],[0.0,0.6,0.8],[0.2,0.8,0.8],[0.425,0.85,0.6375],[0.6375,0.85,0.425],[0.85,0.85,0.22125],[0.95,0.75,0.0]]

plt.figure(figsize=(6, 4), dpi=200)

bp1 = plt.bar(index-bar_width, All, width = bar_width, facecolor = 'lightskyblue', edgecolor = 'black', linestyle=':', label='All')
for patch, color in zip(bp1, colorBar1):
    patch.set_facecolor(color)
    patch.set_alpha(0.3)
bp2 = plt.bar(index, Gain, width = bar_width, facecolor = 'lightskyblue', edgecolor = 'black', linestyle='--', label='Gain')
for patch, color in zip(bp2, colorBar1):
    patch.set_facecolor(color)
    patch.set_alpha(0.6)
bp3 = plt.bar(index+bar_width, Loss, width = bar_width, facecolor = 'lightskyblue', edgecolor = 'black',label='Loss')
for patch, color in zip(bp3, colorBar1):
    patch.set_facecolor(color)
    
plt.xticks(index, celllines, rotation = 0) 
plt.title("RNA Editing Protein Gene Gain and Loss")
plt.ylabel('Protein Gene Number')
plt.legend()
plt.savefig('RNAProteinGeneGainLoss.png')
    

```

```python
input = '''89	143	250
365	762	696
652	1444	954
683	1858	1226
712	1927	1484
626	1743	1048
836	2344	1930
208	500	926'''
celllines = ['MII', 'zygote', 'early', '2cell', '4cell', '8cell', 'ICM', 'mESC']
Noncoding = []
Intro = []
Exon = []

input_process = input.split('\n')

for line in input_process:
    data = line.split('\t')
    Noncoding.append(int(data[0]))
    Intro.append(int(data[1]))
    Exon.append(int(data[2]))

index = np.arange(8)
bar_width = 0.5
plt.figure(figsize=(6, 4), dpi=200)
p1 = plt.bar(index, Noncoding, width = bar_width, facecolor = 'skyblue', edgecolor = 'white', label='Noncoding')
p2 = plt.bar(index, Intro, width = bar_width, bottom=Noncoding, facecolor = 'darkorange', edgecolor = 'white', label='Intro')
p3 = plt.bar(index, Exon, width = bar_width, bottom=Intro, facecolor = 'slategray', edgecolor = 'white', label='Exon')
for p in p1:
    p.set_alpha(0.5)
for p in p2:
    p.set_alpha(0.5)


plt.title('RNA Edit Sites Distribution in Gene')
plt.xticks(index, celllines)
plt.legend()
plt.savefig('RNAEditDistribution.png')
```
